```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

تاریخ_شروع = datetime(2025, 1, 1)
تاریخها = [تاریخ_شروع + timedelta(days=i) for i in range(180)]

محصولات = ['لپ تاپ', 'موبایل', 'تبلت', 'هدفون', 'کیس', 'موس']

دادهها = []
for روز in تاریخها:
    for محصول in محصولات:
        تعداد_فروش = random.randint(0, 20)
        قیمت_واحد = {
            'لپ تاپ': 25000000,
            'موبایل': 15000000,
            'تبلت': 8000000,
            'هدفون': 3000000,
            'کیس': 2000000,
            'موس': 1000000
        }[محصول]
        درآمد = تعداد_فروش * قیمت_واحد
        دادهها.append([روز, محصول, تعداد_فروش, قیمت_واحد, درآمد])

df = pd.DataFrame(دادهها, columns=['تاریخ', 'محصول', 'تعداد فروش', 'قیمت واحد', 'درآمد کل'])
df.head()
```

```python
print(df.groupby('محصول')['تعداد فروش'].sum())
```

```python
print(df.groupby('محصول')['تعداد فروش'].mean())
```

```python
print(df['تعداد فروش'].max())
```

```python
print(df['تعداد فروش'].min())
```

```python
import matplotlib .pyplot as plt 

فروش_هر_محصول=df.groupby('محصول')['تعداد فروش'].sum()

plt.figure(figsize=(10,6))
فروش_هر_محصول.plot(kind='bar', color='skyblue')
plt.title('مجموع فروش هر محصول در 6 ماه')
plt.xlabel('محصول')
plt.ylabel('تعداد فروش')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```
